### Q1
Compute the SIFT features of the two pictures.
a) Extract the SIFT features with default parameters and show the keypoints on the pictures.
Hint: Use existing library functions for this (see suggestions at the end).

```python
import cv2
import matplotlib.pyplot as plt

# read
image1 = cv2.imread("Image1.jpg")
image2 = cv2.imread("Image2.jpg")

# create SIFT
siftDetector = cv2.SIFT_create()

# find keypoints
keypoints1, descriptors1 = siftDetector.detectAndCompute(image1, None)
keypoints2, descriptors2 = siftDetector.detectAndCompute(image2, None)

# draw
image1Keypoint = cv2.drawKeypoints(image1, keypoints1, None)
image2Keypoint = cv2.drawKeypoints(image2, keypoints2, None)

# transform to RGB
image1RGB = cv2.cvtColor(image1Keypoint, cv2.COLOR_BGR2RGB)
image2RGB = cv2.cvtColor(image2Keypoint, cv2.COLOR_BGR2RGB)

# show
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Image1 SIFT keypoints")
plt.imshow(image1RGB)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Image2 SIFT keypoints")
plt.imshow(image2RGB)
plt.axis("off")

plt.show()
```

### Q2
b) To achieve better visualization of the keypoints, reduce their number to include only the
~20 most prominent ones. Hint: Vary the parameter contrastThreshold or nfeatures.
Show the results obtained in a) and b) in your Jupyter notebook (like the examples below) and
include a brief description of the approach you used for b)

```python
# only 20 
siftDetectorB= cv2.SIFT_create(nfeatures=20)

keypoints1B, descriptors1B = siftDetectorB.detectAndCompute(image1, None)
keypoints2B, descriptors2B = siftDetectorB.detectAndCompute(image2, None)

# enlarged size
for kp in keypoints1B:
    kp.size *= 20
for kp in keypoints2B:
    kp.size *= 20

# redraw
image1KeypointB = cv2.drawKeypoints(image1, keypoints1B, None, color=(0,0,255),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
image2KeypointB = cv2.drawKeypoints(image2, keypoints2B, None, color=(0,0,255),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# transform to RGB
image1RGBB = cv2.cvtColor(image1KeypointB, cv2.COLOR_BGR2RGB)
image2RGBB = cv2.cvtColor(image2KeypointB, cv2.COLOR_BGR2RGB)

# show
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Image1 SIFT keypoints")
plt.imshow(image1RGBB)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Image2 SIFT keypoints")
plt.imshow(image2RGBB)
plt.axis("off")

plt.show()
```
